# ***Assignment 4: Aspect opinion***

Once the aspect vocabulary and opinion lexicons are loaded, the opinions about aspects have to be extracted
from the reviews. For this purpose, POS tagging, constituency and dependency parsing could be used.

• POS tagging would allow identifying the adjectives in the sentences.

• Constituency and dependency parsing would allow extracting the relations between nouns and
adjectives and adverbs.

See Appendix C and Appendix D for code examples.

```python
import nltk
nltk.download('averaged_perceptron_tagger')

def pos_tagging(text):
  sentences = nltk.sent_tokenize(text)
  sentences = [nltk.word_tokenize(s) for s in sentences]
  sentences = [nltk.pos_tag(s) for s in sentences]
  return sentences

postagged_sentences = pos_tagging(reviews[0].get("reviewText")) 

print(postagged_sentences[0])
```

### ***Task 4.1 - mandatory:*** 
Extracting the [aspect, aspect term, opinion word, polarity] tuples from the input reviews

```python
# Imports
import json
import numpy as np
import nltk
nltk.download('punkt')

# Structure
children = []

# Open, Load and Save yelp_hotels.json
with open('yelp_dataset/yelp_hotels.json', encoding='utf-8') as f:
    reviews_syn_anal = json.load(f)
f.close()
grammar = r"""
    JJNN: {<JJ>*<NN>+} # chunk adjective and sequences noun
        {<JJ>*<NNP>+} # chunk sequences of proper nouns
"""
cp = nltk.RegexpParser(grammar)

# Save in children structure all the sentences from each review, filtered by words: JJ - adjective & NN/NNP - Noun/Proper Noun
for review in reviews_syn_anal:
    chunk_parse = cp.parse(pos_tagging(review.get("reviewText"))[0])
    for child in chunk_parse:
        if isinstance(child, nltk.Tree): 
            if child.label() == 'JJNN':
                children.append(child)

print("\nEnd\n")                
```

```python
# Entendiendo la estructura generada
print(children[0])
print(children[0][0])
print(children[0][0][0])
print(children[0][0][1])
```

```python
import re

# Find the polarity form each Adjective - jj from the opinion_lexicon Lists (Task 3)
def polarity_finder(jj):
  if(jj in positiveWords): 
    return 1
  elif(jj in negativeWords): 
    return -1
  else:
    return 0

# Find the amenities from a current amenity - Noun - nn
def amenties_finder(nns):
  if(nns in amenity):
    index = amenity.index(nns)
    return amenities[index]

  else:
    return '-'

# Create a new array structure containing dictionaries in form newEntry, Generating the Asked Touples.
def separator(child):
  # Data structures
  clasifier = []
  incomplete = []

  for c in child:
    newEntry = {
        'jj': '',
        'nn': '',
        'pol': 0,
        'amen': '',
    }

    if (len(c) > 1):
      for word in c:
        # Cleaning string input data
        aux = re.sub('[^A-Za-z0-9]+', '', word[0]).lower() 

        # Filtring and organizing words
        if(word[1] == 'JJ'): 
          newEntry['jj'] += aux + ' '
          newEntry['pol'] = polarity_finder(aux)

        if(word[1] == 'NN' or word[1] == 'NNP'):
          newEntry['nn'] += aux + ' '
          newEntry['amen'] = amenties_finder(aux)

      # Processing empty touple nouns or adjective cases
      if(newEntry['jj'] != '' and newEntry['nn'] != ''):
        clasifier.append(newEntry) # Spliting
      else:
        incomplete.append(newEntry) # Spliting

  return clasifier, incomplete

# Use case       
split, incom = separator(children)    
print(*split, sep='\n')
#print(*incom, sep='\n')
```

```python

# Generate new children structure to process the reviews; uses separator function
def process_review(reviewText):
  children2 =  []
  aux = pos_tagging(reviewText)

  for i in range(len(aux)):
      chunk_parse = cp.parse(aux[i])
      for child in chunk_parse:
          if isinstance(child, nltk.Tree): 
              if child.label() == 'JJNN':
                  children2.append(child)

  chil = []
  for c in children2:
    chil.append(c)

  return separator(chil)

# Get all reviews from a given hotel
def review_hotel(asin):
  return [item for item in reviews_syn_anal[:200] if item["asin"] == asin]

# Find all ids for each hotel and save them only once each into a set
def find_all_hotels():
  hotelset = set()
  for hotel in reviews_syn_anal[:200]:
    if hotel['asin'] not in hotelset:
      hotelset.add(hotel['asin'])

  return hotelset
```

```python
# Generates an array in which each position is a different hotel.
# Inside each position we have an array of each reviews for that hotel and the result of processing that review (polarities). 
def group_hotels():
  result = []
  all_hotels = find_all_hotels() #n O(n+m*o)
  for hotel in all_hotels: #m
    hotel_reviews = review_hotel(hotel)
    newHotel = {
      'asin': hotel,
      'reviews': [],
    }
    for entry in hotel_reviews: #o
      a, b = process_review(entry.get("reviewText"))
      newReview = {
          'review': entry.get("reviewText"),
          'value': a
      }
      newHotel['reviews'].append(newReview)

    result.append(newHotel)
  return result
```

```python
processed_review2 = group_hotels()
```

```python
# Coputation for the reviews and the statistics of the total reviews of each hotel.
def calculate_polarity_per_review(i_ran_out_of_names):
  result = []
  for nameless_variable in i_ran_out_of_names:
    newHotelCaculation = {
        'asin': nameless_variable['asin'],
        'reviews': [],
        'stats': {
          'avg': 0,
          'pos': 0,
          'neg': 0,
          'neu': 0
        }
    }
    for corn_kernel in nameless_variable['reviews']:
      if(len(corn_kernel['value']) > 0):

        negative, positive, neutral = 0, 0, 0
        newReviewCaculation = {
          'review': corn_kernel['review'],
          'avg': 0,
          'pos': 0,
          'neg': 0,
          'neu': 0
        }
        for endosperm in corn_kernel['value']:
          ouch = endosperm['pol']
          if (ouch == 1): positive+=1
          if (ouch == 0): neutral+=1
          if (ouch == -1): negative+=1
      
        newReviewCaculation['avg'] = ((negative*-1) + positive) / (negative + positive + neutral)
        newReviewCaculation['pos'] = positive
        newReviewCaculation['neg'] = negative
        newReviewCaculation['neu'] = neutral
        newHotelCaculation['reviews'].append(newReviewCaculation)

      acum = 0
      negative2, positive2, neutral2 = 0, 0, 0

      for reviews in newHotelCaculation['reviews']:
        ouch = reviews['avg']
        acum += ouch
        if (ouch > 0): positive2+=1
        if (ouch == 0): neutral2+=1
        if (ouch < 0): negative2+=1

      newHotelCaculation['stats']['avg'] =  ((negative2*-1) + positive2) / len(nameless_variable['reviews'])
      newHotelCaculation['stats']['pos'] = positive2
      newHotelCaculation['stats']['neg'] = negative2
      newHotelCaculation['stats']['neu'] = neutral2

      result.append(newHotelCaculation)
  return result

calculate_polarity_per_review(processed_review2)
```
